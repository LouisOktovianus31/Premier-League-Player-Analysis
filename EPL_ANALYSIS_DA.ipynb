#**Data Transformation on Highest goal scored by a player in EPL**

```python
max_goal_player = df.loc[df['Goals'].idxmax(), 'Name']
print("Player with the highest goal scored:", max_goal_player)
```

Transformation on Total Goals scored by each club in EPL**

```python
club_goals = df.groupby('Club')['Goals'].sum()
print(club_goals)
```

```python
club_goals_sorted = club_goals.sort_values(ascending=False)
print(club_goals_sorted)
```

```python
import matplotlib.pyplot as plt
```

#**Histogram**

```python

# Create histogram
club_goals_sorted.plot(kind='bar')

# Set plot labels and title
plt.xlabel('Club')
plt.ylabel('Total Goals')
plt.title('Total Goals Scored by Each Club')

# Show the plot
plt.show()
```

#**Boxplot**

```python
# Create boxplot
plt.boxplot(club_goals_sorted.values)

# Set plot labels and title
plt.xticks([1], [club_goals_sorted.index[0]])  # Show only the first club name on x-axis
plt.xlabel('Club')
plt.ylabel('Total Goals')
plt.title('Total Goals Scored by Each Club')

# Show the plot
plt.show()
```

#**Pie Chart**

```python
import matplotlib.pyplot as plt

# Threshold for grouping smaller values
threshold = 2  # 2% threshold

# Find clubs with values under the threshold
small_values = club_goals_sorted[club_goals_sorted < threshold]

# Combine small values into a single category called "Others"
others_value = small_values.sum()
club_goals_combined = club_goals_sorted[club_goals_sorted >= threshold]
club_goals_combined['Others'] = others_value

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(club_goals_combined, labels=club_goals_combined.index, autopct='%1.1f%%', startangle=140)

# Set plot title
plt.title('Distribution of Total Goals by Club')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
